```python
def order_compatibility(order, to_test):
    if not set(to_test.keys()) <= set(order.keys()):
        return False
    for prime in to_test:
        if to_test[prime] > order[prime]:
            return False
    return True
```

```python
def test(k,p): # k points, p prime
    candidate = 0
    while k>p-1:
        base_candidate_power = floor(log(k,p)+1) # log(p*k,p) = 1+log(k,p)
        base_candidate_points = (p**base_candidate_power-1)/(p-1)
        # Points needed for promotion = base_candidate_power
        if base_candidate_points - base_candidate_power > k:
            # We are not in promotion range
            base_candidate_power -= 1
            base_candidate_points = (p**base_candidate_power-1)/(p-1)
            k -= base_candidate_points
            candidate += p**base_candidate_power
            # print(k, candidate)
        elif base_candidate_points > k:
            # We are in promotion range but do not have enough to promote
            # We add until previous to promotion and remove points
            candidate += p**base_candidate_power - p
            k = 0
        else:
            # We promote and remove points
            k -= base_candidate_points
            candidate += p**base_candidate_power
            # print(k, candidate)
    # Extend until just befote the next multiple + excess
    candidate += p*k + p-1
    return candidate
```

```python
def clean_test(k,p):
    candidate = 0
    while k>p-1:
        base_candidate_power = floor(log(k,p)+1) # log(p*k,p) = 1+log(k,p)
        base_candidate_points = (p**base_candidate_power-1)/(p-1)
        # Points needed for promotion = base_candidate_power
        if base_candidate_points - base_candidate_power > k:
            # We are not in promotion range
            base_candidate_power -= 1
            k -= (p**base_candidate_power-1)/(p-1)
            candidate += p**base_candidate_power
        elif base_candidate_points > k:
            # We are in promotion range but do not have enough to promote
            # We add until previous to promotion and remove points
            candidate += p**base_candidate_power - p
            k = 0
        else:
            # We promote and remove points
            k -= base_candidate_points
            candidate += p**base_candidate_power
    # Extend until just befote the next multiple + excess points
    candidate += p*k + p-1
    return candidate
```

## PGL 12-19

```python
for i in range(12,21):
    orders = build_bound(i)
    bound_order = orders
    boundSimpleGroups = SimpleGroupsOrder(multipliers=True)
    boundSimpleGroups.set_order(bound_order)
    boundSimpleGroups.full_check()
    print('PGL' + str(i) + '_autobuild.txt')
    print('Trivial multipliers')
    print([g for g in boundSimpleGroups.get_calculated_groups() if boundSimpleGroups.get_calculated_groups()[g] == 1])
    print('Non-trivial multipliers')
    print([g for g in boundSimpleGroups.get_calculated_groups() if boundSimpleGroups.get_calculated_groups()[g] != 1])
    boundSimpleGroups.standard_notation()
    report(boundSimpleGroups, 'PGL_autobuilds/PGL' + str(i) + '_autobuild.txt')
```

### Bound tables

```python
max_dim = 32
primes = list(build_bound(max_dim).keys())
bounds = [build_bound(i) for i in range(2,max_dim+1)]
bound_table = "n"
for p in primes:
    bound_table += " & " + str(p)
bound_table += " \\\\ \\hline \n"
j = 2
for bound in bounds:
    bound_table += str(j)
    for p in primes:
        if p in bound:
            bound_table += " & " + str(bound[p])
        else:
            bound_table += " & 0"
    bound_table += " \\\\ \\hline \n"
    j += 1
```

```python
print(bound_table)
```
